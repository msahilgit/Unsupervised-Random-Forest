```python
import copy as cp
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

```python
paths = [
    'saved_dt1/',
    'saved_dt2/',
    'saved_dt4/',
    'saved_dt5/'
]
```

```python
losses = [np.load(f'{p}loss_train.npy') for p in paths]
vlosses = [np.load(f'{p}loss_valid.npy') for p in paths]
eigs = [np.load(f'{p}evals_train.npy') for p in paths]
```

```python
nn=4

fig,ax = plt.subplots(1,figsize=(5,6))
for spine in ax.spines.values(): spine.set_linewidth(2)
ax.set_xlim(-2,30)
ax.set_ylim(-2.1,-0.7)
ax.set_xticks([0,15,30])
ax.set_yticks([-2.1,-1.4,-0.7])
ax.tick_params(length=8, width=2, labelsize=24)
ax.xaxis.set_minor_locator(AutoMinorLocator(3))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(which='minor', length=5, width=2)
#
ax.plot(losses[nn-1], lw=2, color='c')
ax.plot(vlosses[nn-1], lw=2, color='y')
#
plt.savefig(f'saved_plots/loss{nn}.png', transparent=True)
plt.show()
```
