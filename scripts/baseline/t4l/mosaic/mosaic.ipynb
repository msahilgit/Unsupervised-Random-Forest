```python
import copy as cp
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

## t4l

```python
cmat = [np.load(f'saved_t4l_mosaic/corr_mat{i}.npy') for i in range(5)]
```

```python
clusters = [np.load(f'saved_t4l_mosaic/clusters_{i}.npy', allow_pickle=True) for i in range(5)]
```

```python
i=4

idxs = np.argsort(
    [len(cluster) for cluster in clusters[i]],
)[::-1]
clusters_sorted = clusters[i][idxs]
clusters_sorted_flattened = np.concatenate(clusters[i][idxs])

# sort the matrix accordingly
matrix_sorted = cmat[i][
    np.ix_(clusters_sorted_flattened, clusters_sorted_flattened)
]
```

```python
lengths = np.array([len(cluster) for cluster in clusters_sorted])
last5 = np.where(lengths>=5)[0][-1]
ticks = np.cumsum(lengths[:last5+1])
```

```python
fig, ax = plt.subplots(1, figsize=(5,6))
im = ax.pcolormesh(
    matrix_sorted,
    snap=True,
    vmin=0,
    vmax=1,
    cmap='jet'
)
ax.invert_yaxis()  
ax.set_aspect('equal')
ax.set_xticks(ticks)  
ax.set_yticks(ticks)
ax.tick_params(labelleft=False, labelbottom=False)
ax.grid(True)
#
plt.savefig(f'saved_plots/t4l_clusters{i}.png', transparent=True)
plt.show()
```
