```python
import copy as cp
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

```python
kloss = [
    [
        np.load(f'saved_tvae{nn}/kl_loss_{i}.npy')
    for i in range(5)]
for nn in range(1,6)]
```

```python
rloss = [
    [
        np.load(f'saved_tvae{nn}/rc_loss_{i}.npy')
    for i in range(5)]
for nn in range(1,6)]
```

```python
colors = ['green', 'darkturquoise', 'indianred', 'blueviolet', 'gray']
```

```python
nn=5

fig,ax = plt.subplots(1,figsize=(5,6))
ax.spines[:].set_linewidth(2)
ax.set_xlim(-2,41)
ax.set_ylim(-0.5,10)
ax.set_xticks([0,20,40])
ax.set_yticks([0,5,10])
ax.tick_params(length=8, width=2, labelsize=24)
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(3))
ax.tick_params(which='minor', length=5, width=2)
#
for a,loss in enumerate(kloss[nn-1]): ax.plot(loss*1000, lw=2, color=colors[a])
#
plt.savefig(f'saved_plots/kloss{nn}.png', transparent=True)
plt.show()
```
