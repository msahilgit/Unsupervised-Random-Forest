```python
import copy as cp
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

```python
colors = ['green', 'darkturquoise', 'indianred', 'blueviolet', 'gray']
```

```python
paths = [
    'saved_relu_ae1/',
    'saved_relu_ae2/',
    'saved_relu_ae3/',
    'saved_relu_ae4/',
    'saved_tanh_ae4/'
]
```

```python
losses = [
    [
        np.load(f'{p}loss_{i}.npy')
    for i in range(5)]
for p in paths]
```

```python
vlosses = [
    [
        np.load(f'{p}val_loss_{i}.npy')
    for i in range(5)]
for p in paths]
```

```python
nn=5

fig,ax = plt.subplots(1,figsize=(5,6))
ax.spines[:].set_linewidth(2)
ax.set_xlim(-5,201)
ax.set_ylim(2.5,6)
ax.set_xticks([0,100,200])
ax.set_yticks([3,4,5,6])
ax.tick_params(length=8, width=2, labelsize=24)
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(which='minor', length=5, width=2)
#
for a,loss in enumerate(losses[nn-1]): ax.plot(loss*100, lw=2, color=colors[a])
#
plt.savefig(f'saved_plots/loss{nn}.png', transparent=True)
plt.show()
```
