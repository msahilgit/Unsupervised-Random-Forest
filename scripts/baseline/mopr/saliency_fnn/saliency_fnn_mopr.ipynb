```python
import copy as cp
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

```python
loss = [
    [
        np.load(f'saved_nn{n}/loss_{i}.npy')
    for i in range(5)]
for n in range(1,7)]
```

```python
colors = ['green', 'darkturquoise', 'indianred', 'blueviolet', 'gray']
```

```python
nn=6

fig,ax = plt.subplots(1,figsize=(5,6))
ax.spines[:].set_linewidth(2)
ax.set_xlim(-5,201)
ax.set_ylim(0,0.01)
ax.set_xticks([0,100,200])
ax.set_yticks([0,0.01])
ax.tick_params(length=8, width=2, labelsize=24)
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))
ax.tick_params(which='minor', length=5, width=2)
#
for a,curve in enumerate(loss[nn-1]): ax.plot(curve, lw=2, color=colors[a])
#
plt.savefig(f'saved_plots/loss_nn{nn}.png', transparent=True)
plt.show()
```
