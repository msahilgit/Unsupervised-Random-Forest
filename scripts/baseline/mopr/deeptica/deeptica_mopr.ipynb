```python
import copy as cp
import numpy as np
import matplotlib as mt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

```python
loss = [np.load(f'saved_dt{nn}/loss_train.npy') for nn in range(1,4)]
vloss = [np.load(f'saved_dt{nn}/loss_valid.npy') for nn in range(1,4)]
```

```python
eig = [np.load(f'saved_dt{nn}/evals_train.npy') for nn in range(1,4)]
```

```python
nn=1

fig,ax = plt.subplots(1,figsize=(5,6))
for spine in ax.spines.values(): spine.set_linewidth(2)
ax.set_xlim(-1,14)
ax.set_ylim(-1.902,-1.8)
ax.set_xticks([0,7,14])
ax.set_yticks([-1.9,-1.8])
ax.tick_params(length=8, width=2, labelsize=24)
ax.xaxis.set_minor_locator(AutoMinorLocator(2))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(which='minor', length=5, width=2)
#
ax.plot(loss[nn-1], lw=2, color='c')
ax.plot(vloss[nn-1], lw=2, color='y')
#
plt.savefig(f'saved_plots/loss{nn}.png', transparent=True)
plt.show()
```
