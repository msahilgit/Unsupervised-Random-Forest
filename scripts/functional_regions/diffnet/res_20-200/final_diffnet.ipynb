```python
import numpy as np
import os
# from diffnets import utils
import pickle
from collections import defaultdict
import matplotlib.pyplot as plt
import matplotlib as mt
from matplotlib.ticker import AutoMinorLocator, MaxNLocator
```

```python
whitened_dir = "2_processed_data/whitened_data"
```

### 64

```python
nn_dir = '3_ss_sae_1/latent64_0111/'
```

```python
rmsd = np.load(f'{nn_dir}/rmsd.npy')
```

```python
bins = np.arange(rmsd.min(), rmsd.max(), 0.04)
xb = bins[1:] - (bins[1]-bins[0])/2
zeros = np.zeros(xb.shape)

hrmsd = np.histogram(rmsd, bins=bins, density=False)[0]/rmsd.shape[0]
```

```python
fig,ax=plt.subplots(figsize=(8,6))
ax.plot(xb, hrmsd, color='crimson', lw=2)
ax.fill_between(xb, hrmsd, zeros, color='crimson', alpha=0.5)
#
ax.set_xlim(0,1.2)
ax.set_ylim(0,0.16)
ax.set_xticks([0,0.4, 0.8, 1.2])
ax.set_yticks([0,0.08,0.16])
for spine in ax.spines.values(): spine.set_linewidth(2)
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.tick_params(labelsize=22, length=7, width=2, )
ax.tick_params(which='minor', length=4, width=2)
plt.savefig(f'plots/rmsd64.pdf')
plt.show()
```

```python
output_labels = [np.load(f'{nn_dir}/labels/lab00000{i}.npy')[:,0] for i in range(4)]
```

```python
bins = np.arange(0,1.01,0.01)
xb = bins[1:]-(bins[1]-bins[0])/2

houtput_labels = [np.histogram(ol, bins=bins, density=False)[0]/ol.shape[0] for ol in output_labels]
```

```python
colors = ['red', 'blue', 'green', 'gray']
```

```python
fig,ax=plt.subplots(figsize=(8,6))
for a, hh in enumerate(houtput_labels):
    ax.plot(xb, hh, color=colors[a], lw=8, alpha=0.7)
ax.set_xlim(-0.02,1.02)
ax.set_ylim(-0.02,0.85)
ax.set_yticks([0,0.4, 0.8])
ax.set_xticks([0,0.5,1])
for spine in ax.spines.values(): spine.set_linewidth(2)
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.tick_params(labelsize=22, length=7, width=2, )
ax.tick_params(which='minor', length=4, width=2)
plt.savefig(f'plots/output_labels64.pdf')
plt.show()
```
